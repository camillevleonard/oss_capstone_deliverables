### Proof of Concept Calls 
#### inspirare Information

```python
r_inspirare = requests.get('https://try.gitea.io/api/v1/orgs/inspirare')
r_inspirare.text
r_inspirare_json = json.loads(r_inspirare.text)
r_inspirare_df = pd.json_normalize(r_inspirare_json)
r_inspirare_df
```

#### inspirare Public Members

```python
r_inspirare_pub_mem = requests.get('https://try.gitea.io/api/v1/orgs/inspirare/members?page=1')
r_inspirare_pub_mem.text
r_inspirare_pub_mem_json = json.loads(r_inspirare_pub_mem.text)
r_inspirare_pub_mem_df = pd.json_normalize(r_inspirare_pub_mem_json)
r_inspirare_pub_mem_df
```

#### inspirare Public Repos 

```python
r_inspirare_repos = requests.get('https://try.gitea.io/api/v1/orgs/inspirare/repos') 
r_inspirare_repos.text
r_inspirare_repos_json = json.loads(r_inspirare_repos.text)
r_inspirare_repos_df = pd.json_normalize(r_inspirare_repos_json)
r_inspirare_repos_df
```

#### Additional Calls

* /orgs/{org}/labels list an organization's labels
* /orgs/{org}/members list an organization's members 
* /orgs/{org}/public_members list an organizations public members 
* /orgs/{org}/public_members/{username} check if a user is a public member of an org
* orgs/{org}/repos list an organization's repos 
* orgs/{org}/teams list an organization's teams 
* teams/{is}/members list a team's members 
* teams/{id}/repos list a team's repos

### Get Repos - It appears you cannot access a list of repos 

```python
r_repos = requests.get('https://try.gitea.io/api/v1/repos/search')
r_repos_json = json.loads(r_repos.text)
r_repos_df = pd.DataFrame()

len(r_repos_json.get('data'))
```

```python
for i in range(0,30):
   r_repos_df = r_repos_df.append(pd.json_normalize(r_repos_json.get('data')[i]) )
```
